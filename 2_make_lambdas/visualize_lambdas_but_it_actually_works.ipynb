```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
path = "/Users/douglas/Documents/Features_Transfer/store_lambdas/"
savepath = "/Users/douglas/Documents/Features_Transfer/figures/"
```

```python
directory_numbers = [413,414,416,417,419,420,421,422,423,424,425]
for i in directory_numbers:
    Lambdas = np.loadtxt(path+f'lambdas{i}.txt',delimiter=",")
    plt.matshow(Lambdas,vmin=Lambdas[0,2],vmax =Lambdas[0,4999],cmap='Blues_r') 
    plt.colorbar()
    plt.savefig(savepath+f'lambdas{i}.svg')
    plt.title(f'lambdas{i}')
    plt.show()
```
