```python
import matplotlib as mpl
import matplotlib.pyplot as plt #pyplot is a sub-module of matplotlib which doesn't get imported with a simple import matplotlib.
import numpy as np
```

```python
path = "/Users/douglas/Documents/Features Transfer/store lambdas/"
savepath = "/Users/douglas/Documents/Features Transfer/figures/"
directory = 8 #directory number
Lambdas = np.loadtxt(path+"lambdas"+str(directory)+".txt",delimiter=",")
```

```python
# setup to plot lambdas

# interaction alphas from the MiChroM for a couple type-to-type interactions
A1B1 = -0.262513 #smallest magnitude, but farthest right on a number line
A1A1 = -0.268028
B1B1 = -0.342020
vmax = A1B1
vmin = B1B1

#Colormaps map the data to be plotted into the range [0,1]. I need the mapped number associated with A1A1.
special_number_for_A1A1 = (A1A1-vmin)/(vmax-vmin)
print("special_number_for_A1A1 =",special_number_for_A1A1)

nodes = [0,special_number_for_A1A1,1]
```

```python
# Plot lambdas

# Make a colormap to distinguish the three type-to-type interactions by color.
colors = ['dimgray','lightgray','white']
my_cmap = mpl.colors.LinearSegmentedColormap.from_list("my_cmap", list(zip(nodes,colors)))

for spine in plt.gca().spines.values():#Removes the frame box
    spine.set_visible(False)

plt.box('False')
plt.imshow(Lambdas,vmin=vmin,vmax=vmax,cmap=my_cmap) 
plt.colorbar()
plt.tick_params(left = False, right = False , labelleft = False , labelbottom = False, bottom = False)#Remove ticks and axis labels.
plt.savefig(savepath+"lambdas"+str(directory)+".pdf")
plt.show()
```

```python
# Plot lambdas

# Make a colormap that doesn't distinguish the three type-to-type interactions by color.
colors = ['royalblue','white']
my_cmap = mpl.colors.LinearSegmentedColormap.from_list("my_cmap", list(zip([0,1],colors)))

for spine in plt.gca().spines.values():#Removes the frame box
    spine.set_visible(False)
    
plt.imshow(Lambdas,vmin=-0.29,vmax =vmax,cmap=my_cmap) 
plt.colorbar()
plt.tick_params(left = False, right = False , labelleft = False , labelbottom = False, bottom = False)#Remove ticks and axis labels.
plt.savefig(savepath+"lambdas"+str(directory)+".pdf")
plt.show()
```
