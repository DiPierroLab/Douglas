```python
import matplotlib.pyplot as plt
import numpy as np
```

### Fits to MiChroM optimization from MiChroM supplement

```python
# Translate the above functions into python
def gamma_exp(d, A=1.969, B=2.433, C=-.015):
    return -A*np.exp(-d/B)+C

def gamma_2(d, A=0.056, B= 4.764):
    return -A/np.log(d)-B/d**2

def gamma_3(d, A=0.030, B=0.351, C=3.727):
    return -A/np.log(d)-B/d-C/d**2
```

### Reparametrize fits to make gammas

```python
d = np.array(range(2500))
bead_size = .5 # kilobases
```

```python
i = 112 # number of first
for A in [.25,.3]:
    for B in [800,1200,2000,3000]:
        for C in [0.0]:
            gamma_i = gamma_exp(d,A,B,C)
            np.savetxt(f'gamma_{i}.txt',gamma_i)
            plt.plot(bead_size*d,gamma_i,label=f'sim {i}  A={A}  B={B}  C={C}')
            i += 1
plt.legend(title = 'A e **(-d/B)+C')
plt.xlabel('genomic distance d')
plt.ylabel('gamma')
plt.show()      
```

```python
gamma_122_cis = gamma_exp(d,A=.2,B=1600,C=0.0)
gamma_122_trans = gamma_exp(d,A=.2,B=400,C=0.0)
np.savetxt('gamma_122_cis.txt',gamma_122_cis)
np.savetxt('gamma_122_trans.txt',gamma_122_trans)
plt.plot(bead_size*d,gamma_122_cis,label='sim_122_cis  A=.2  B=1600  C=0')
plt.plot(bead_size*d,gamma_122_trans,label='sim_122_trans  A=.2  B=400  C=0')
plt.legend(title = 'A e **(-d/B)+C')
plt.xlabel('genomic distance d')
plt.ylabel('gamma')
plt.show()  
```

```python
i = 123 # number of first
A = .2
B_cis = 800
C = 0.0
gamma_123_to_126_cis = gamma_exp(d,A,B_cis,C)
np.savetxt('gamma_123_to_126_cis.txt',gamma_123_to_126_cis)
plt.plot(bead_size*d,gamma_123_to_126_cis,label=f'sim 123-126 cis   A={A}  B={B_cis}  C={C}')
for B_trans in [200,100,50,25]:
    gamma_i_trans = gamma_exp(d,A,B_trans,C)
    np.savetxt(f'gamma_{i}_trans.txt',gamma_i_trans)
    plt.plot(bead_size*d,gamma_i_trans,label=f'sim {i} trans   A={A}  B={B_trans}  C={C}')
    i += 1
plt.plot()
plt.legend(title = 'A e **(-d/B)+C')
plt.xlabel('genomic distance d (kb)')
plt.xlim(0,400)
plt.ylabel('gamma')
plt.show()
```
