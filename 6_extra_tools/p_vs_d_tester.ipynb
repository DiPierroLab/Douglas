```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
pathHiC = "/Users/douglas/Documents/Features Transfer/Hi-C/"
M69 = np.loadtxt(pathHiC+"Hi-C_directory_69.txt")
M70 = np.loadtxt(pathHiC+"Hi-C_directory_70.txt")
M69block11 = M69[0:2500,0:2500]
M70block11 = M70[0:2500,0:2500]
M69block22 = M69[2500:5000,2500:5000]
M70block22 = M70[2500:5000,2500:5000]
d = np.arange(0,2500,1)
```

```python
pathSeq = "/Users/douglas/Documents/Features Transfer/sequences/store sequences/"
seqPat = np.loadtxt(pathSeq+"chr_AAAA_2500_beads.txt",str)#sequence of maternal chromosome
seqMat = np.loadtxt(pathSeq+"chr_ABAB_2500_beads.txt",str)#sequence of maternal chromosome
```

```python
def Ps(M):
    N = M.shape[0]
    PsArray = np.zeros(N)
    for d in range(N):
        Ps = np.diag(M,k=d) # kth off diagonal of the input matrix M (0th = diagonal)
        avg = np.average(Ps) # average contact probability at genomic distance d
        PsArray[d] = avg
    return PsArray
```

```python
def PsType(M,seq1,seq2,type1,type2):
    N = M.shape[0]
    PsArray = np.zeros(N)
    for d in range(N):
        Ps = np.diag(M,k=d) # kth off diagonal of the input matrix M (0th = diagonal)
        totalP = 0.0
        counter = 0
        for i in range(N-d):
            j = i + d
            if (seq1[i,1]==type1 and seq2[j,1]==type2) or (seq1[i,1]==type2 and seq2[j,1]==type1):
                totalP += M[i,j]
                counter += 1
        avg = 0.0
        if counter!=0:
            avg = totalP/counter # average contact probability at genomic distance d
        PsArray[d] = avg
    return PsArray
```

```python
#Test that PsType(.) bins AA, AB, and BB correctly  and  divides by the correct number for the various types. 
d = np.loadtxt("/Users/douglas/Documents/Features Transfer/Ps vs Genomic graphs/d.txt")

#Make test contact map
Mtest = np.zeros([2500,2500])#a fake Hi-C map
for i in range(2500):
    for j in range(2500):
        if (seqMat[i,1]=='A1' and seqMat[j,1]=='B1') or (seqMat[i,1]=='B1' and seqMat[j,1]=='A1'):
            Mtest[i,j]=1
        if seqMat[i,1]=='A1' and seqMat[j,1]=='A1':
            Mtest[i,j]=2
        if seqMat[i,1]=='B1' and seqMat[j,1]=='B1':
            Mtest[i,j]=3
plt.imshow(Mtest)
plt.colorbar()
plt.title("Fake contact map ABAB cis for testing")
plt.show()

#plot tests
PsTest = Ps(Mtest)
plt.plot(d,PsTest,label='Ps Test',color='grey')
PsTestAB = PsType(Mtest,seqMat,seqMat,'A1','B1')
plt.plot(d,PsTestAB,label='Ps Test AB',color='navy')
PsTestAA = PsType(Mtest,seqMat,seqMat,'A1','A1')
plt.plot(d,PsTestAA,label='Ps Test AA',color='teal')
PsTestBB = PsType(Mtest,seqMat,seqMat,'B1','B1')
plt.plot(d,PsTestBB,label='Ps Test BB',color='gold')
plt.legend()
plt.show()
```
