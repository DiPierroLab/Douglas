```python
import matplotlib.pyplot as plt
import numpy as np

path = "/Users/douglas/Documents/Features Transfer/store_Ps_vs_genomic_graphs/"
```

```python
#Load P vs d data from directories 10 and 11 into numpy arrays.
#Directories 10 and 11 both have the sequence AAAA ABAB, and 11 has a trans ideal chromosome.

d = np.loadtxt(path+"d.txt")

dir10_block22_P = np.loadtxt(path+"PvsGenomic_directory_10.txt")
dir10_block22_PAA = np.loadtxt(path+"PAAvsGenomic_directory_10.txt")
dir10_block22_PAB = np.loadtxt(path+"PABvsGenomic_directory_10.txt")
dir10_block22_PBB = np.loadtxt(path+"PBBvsGenomic_directory_10.txt")

dir11_block22_P = np.loadtxt(path+"PvsGenomic_directory_11.txt")
dir11_block22_PAA = np.loadtxt(path+"PAAvsGenomic_directory_11.txt")
dir11_block22_PAB = np.loadtxt(path+"PABvsGenomic_directory_11.txt")
dir11_block22_PBB = np.loadtxt(path+"PBBvsGenomic_directory_11.txt")
```

```python
# plot data 10

#plt.plot(d,dir10_block22_P,color='grey', label='P')
plt.plot(d,dir10_block22_PAA,color='yellowgreen', label="PAA")
plt.plot(d,dir10_block22_PAB,color='orange', label="PAB")
plt.plot(d,dir10_block22_PBB,color='teal', label='PBB')

plt.xlabel('d')
plt.ylabel('P(d)')
plt.legend()
plt.loglog()
#plt.axvline(x = 625, color = 'black')
#plt.axvline(x = 1250, color = 'black')
#plt.axvline(x = 1875, color = 'black')
plt.title('P vs d for directory 10 (without trans IC)')
plt.savefig(path+"P_vs_d_10.svg")
```

```python
# plot data 11

#plt.plot(d,dir11_block22_P,color='grey', label='P')
plt.plot(d,dir11_block22_PAA,color='yellowgreen', label="PAA")
plt.plot(d,dir11_block22_PAB,color='orange', label="PAB")
plt.plot(d,dir11_block22_PBB,color='teal', label="PBB")

plt.xlabel('d')
plt.ylabel('P')
plt.legend()
#plt.xlim([-1,5])
plt.loglog()
#plt.axvline(x = 625, color = 'black')
#plt.axvline(x = 1250, color = 'black')
#plt.axvline(x = 1875, color = 'black')
plt.title("P vs d for directory 11 (with trans IC)")
plt.savefig(path+"P_vs_d_11.svg")
```

```python
# plot PBB/PAA to compare dir10 and dir11

plt.plot(d,dir10_block22_PBB/dir10_block22_PAA,label="PBB/PAA no trans IC")
plt.plot(d,dir11_block22_PBB/dir11_block22_PAA,label="PBB/PAA with trans IC")

plt.xlabel('d')
plt.ylabel('PBB/PAA')
plt.legend()
plt.semilogx()
#plt.axvline(x = 625, color = 'black')
#plt.axvline(x = 1250, color = 'black')
#plt.axvline(x = 1875, color = 'black')
plt.title("PBB/PAA vs d")
plt.savefig(path+'PBB_over_PAA_vs_d.svg')
```

```python
# plot PBB-PAA to compare dir10 and dir11

plt.plot(d,dir10_block22_PBB-dir10_block22_PAA,label="PBB-PAA no trans IC")
plt.plot(d,dir11_block22_PBB-dir11_block22_PAA,label="PBB-PAA with trans IC")

plt.xlabel('d')
plt.ylabel('PBB-PAA')
plt.legend()
plt.loglog()
#plt.axvline(x = 625, color = 'black')
#plt.axvline(x = 1250, color = 'black')
#plt.axvline(x = 1875, color = 'black')
plt.title("PBB-PAA vs d")
plt.savefig(path+'PBB_minus_PAA_vs_d.svg')
```

```python
# P11/P10

#plt.plot(d,dir11_block22_P/dir10_block22_P,color='grey', label='P11/P10')
plt.plot(d,dir11_block22_PAA/dir10_block22_PAA,color='yellowgreen',label="PAA11/PAA10")#with trans IC divided by without trans IC
plt.plot(d,dir11_block22_PAB/dir10_block22_PAB,color='orange', label="PAB11/PAB10")
plt.plot(d,dir11_block22_PBB/dir10_block22_PBB,color='teal', label="PBB11/PBB10")

plt.xlabel('d')
plt.ylabel('P11/P10')
plt.legend()
plt.semilogx()
#plt.axvline(x = 625, color = 'black')
#plt.axvline(x = 1250, color = 'black')
#plt.axvline(x = 1875, color = 'black')
plt.title("P11/P10 vs d")
plt.savefig(path+'P11_over_P10_vs_d.svg')
```
