```python
import matplotlib.pyplot as plt
import numpy as np
import cooler
```

```python
# coarse grains an input square matrix
def coarse_grainer(input_matrix, res):
    # res is the number of old bins per new bin
    N_input = input_matrix.shape[0]
    N_coarse = int(N_input/res)
    output_matrix = np.zeros(N_coarse)
    for i in range(N_coarse):
        output_matrix[i] = np.average(input_matrix[i*res:(i+1)*res])
    return output_matrix
```

```python
# Compute experimental contact frequency vs genomic distance at 1kb resolution

# position of a loose region along the trans diagonal
start_index = 6120 # kb
stop_index = 6330 # kb

# Load and prepare Hi-C matrix
clr_1kb = cooler.Cooler('/Users/douglas/Documents/Features_Transfer/store_Hi-C/JJ10_11_17_18.wo3kbCisTransHom.1000.cool')
trans_matrix = clr_1kb.matrix(balance=False)[start_index:stop_index, start_index + 44159: stop_index + 44159]
trans_matrix = np.nan_to_num(trans_matrix) # make the nan entries 0

exp_avg_loose = np.average(trans_matrix)
print('exp_avg_loose =',exp_avg_loose)
```

```python
# Find average contact probability in trans within the loosely paired region of a simulation
HiC_sim = np.loadtxt('/Users/douglas/Documents/Features_Transfer/store_Hi-C/Hi-C_directory_409.txt')
sim_avg_loose = np.average(HiC_sim[1000:1500,3500:4000])
```

```python
# Load probability or contact frequency vs genomic distance.
P_exp = np.loadtxt('/Users/douglas/Documents/Features_Transfer/store_Ps_vs_genomic_graphs/P_exp_chr2_1kb.txt')
P_sim = np.loadtxt('/Users/douglas/Documents/Features_Transfer/store_Ps_vs_genomic_graphs/PvsGenomic_directory_409.txt')

# Rebin
P_sim = coarse_grainer(P_sim,2)

# Rescale
pos = 5 # bin at which to equate cis experiment and cis simulation
rescale_factor = P_exp[pos]/P_sim[pos]
P_sim *= rescale_factor
sim_avg_loose *= rescale_factor
print('sim_avg_loose =',sim_avg_loose)

# Make genomic distance array 
d = np.array(range(1250)) # kilobases
```

```python
# plot experiment
plt.plot(d, P_exp[0:1250], label='exp', color = 'b')
plt.axhline(y = exp_avg_loose, color = 'b', linestyle = 'dashed', label='exp avg loose contact freq') 
plt.axvline(x = 35, color = 'b', linestyle = 'dotted',label='exp tight = loose genomic distance') 

# plot simulation
plt.plot(d, P_sim, label='sim', color = 'r')
plt.axhline(y = sim_avg_loose, color = 'r', linestyle = 'dashed', label='sim avg loose contact freq')
plt.axvline(x = 26, color = 'r', linestyle = 'dotted', label='sim tight = loose genomic distance') 

plt.xlabel('genomic distance (kb)')
plt.xlim(4,50)
#plt.xscale('log')
plt.ylabel('contact frequency')
plt.ylim(3e-3,3)
#plt.yscale('log')
plt.legend()
plt.show()
```
