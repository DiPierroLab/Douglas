```python
Ps = [] #contact probability as a function of genomic distance (Genomic distance = index in this case.)
ds = range(N)
for d in range(N):
    P_d = (np.average(np.diag(HiC,k=d))+np.average(np.diag(HiC,k=-d)))/2
    Ps.append(P_d)
```
