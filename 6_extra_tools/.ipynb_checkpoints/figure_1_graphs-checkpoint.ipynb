```python
import numpy as np
import matplotlib.pyplot as plt
path = '/Users/douglas/Documents/Features Transfer/figures/'
```

```python
mu = 3.22 #same as in Supplementary Materials for A Transferable Model For Chromosome Architecture
r_c = 1.78 #same as in Supplementary Materials for A Transferable Model For Chromosome Architecture
def f(r_ij):
    return .5*(1+np.tanh(mu*(r_c-r_ij)))

x_f = np.linspace(0,2*r_c,100)
y_f = f(x_f)

plt.plot(x_f,y_f)
plt.savefig(path+'f_sigmoid.pdf',format='pdf')
plt.show()
```

```python
def gamma(d): # d represents the trans homolog genomic distance or the genomic distance.
    gamma_1 = -0.030
    gamma_2 = -0.351
    gamma_3 = -3.727
    return gamma_1/np.log(d)+gamma_2/d+gamma_3/d**2

x_gamma = np.linspace(2,100,100)
y_gamma = gamma(x_gamma)

plt.plot(x_gamma,y_gamma)
plt.savefig(path+'gamma.pdf',format='pdf')
plt.show()
```
