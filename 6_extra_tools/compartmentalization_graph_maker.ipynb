```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
M_unpaired = np.loadtxt('/Users/douglas/Documents/Features_Transfer/store_Hi-C/Hi-C_directory_425.txt')
M_paired = np.loadtxt('/Users/douglas/Documents/Features_Transfer/store_Hi-C/Hi-C_directory_423.txt')
```

```python
plt.matshow(M_unpaired, norm=mpl.colors.LogNorm(vmin=0.00003, vmax=1.0),cmap="Reds")
plt.title('Unpaired')
plt.colorbar()
```

```python
PAA_unpaired = np.loadtxt('/Users/douglas/Documents/Features_Transfer/store_Ps_vs_genomic_graphs/PAAvsGenomic_directory_425.txt')
PAB_unpaired = np.loadtxt('/Users/douglas/Documents/Features_Transfer/store_Ps_vs_genomic_graphs/PABvsGenomic_directory_425.txt')
PBB_unpaired = np.loadtxt('/Users/douglas/Documents/Features_Transfer/store_Ps_vs_genomic_graphs/PBBvsGenomic_directory_425.txt')

PAA_paired = np.loadtxt('/Users/douglas/Documents/Features_Transfer/store_Ps_vs_genomic_graphs/PAAvsGenomic_directory_423.txt')
PAB_paired = np.loadtxt('/Users/douglas/Documents/Features_Transfer/store_Ps_vs_genomic_graphs/PABvsGenomic_directory_423.txt')
PBB_paired = np.loadtxt('/Users/douglas/Documents/Features_Transfer/store_Ps_vs_genomic_graphs/PBBvsGenomic_directory_423.txt')
```

```python
d = np.arange(2500)*.5

plt.plot(d,PAB_unpaired, label='PAB_unpaired')
plt.plot(d,PBB_unpaired, label='PBB_unpaired')
plt.plot(d,PAB_paired, label='PAB_paired')
plt.plot(d,PBB_paired, label='PBB_paired')

plt.legend()
plt.xlabel('genomic distance from diagonal (kb)')
plt.xlim((1,200))
plt.ylabel('relative contact frequency')
plt.yscale('log')
plt.show()
```

```python
d = np.arange(2500)#*.5

plt.plot(d,(PAA_unpaired+PBB_unpaired)/(2*PAB_unpaired), label='(PAA_unpaired+PBB_unpaired)/(2*PAB_unpaired)')
plt.plot(d,(PAA_paired+PBB_paired)/(2*PAB_paired), label='(PAA_paired+PBB_paired)/(2*PAB_paired)')

plt.legend()
plt.xlabel('genomic distance from diagonal (kb)')
plt.xlim((1,200))
plt.ylabel('relative contact frequency')
plt.show()
```

```python
d = np.arange(2500)#*.5

plt.plot(d,PAB_unpaired/PBB_unpaired, label='PAB_unpaired/PBB_unpaired')
plt.plot(d,PAB_paired/PBB_paired, label='PAB_paired/PBB_paired')

plt.legend()
plt.xlabel('genomic distance from diagonal (kb)')
plt.xlim((1,200))
plt.ylabel('relative contact frequency')
plt.show()
```
